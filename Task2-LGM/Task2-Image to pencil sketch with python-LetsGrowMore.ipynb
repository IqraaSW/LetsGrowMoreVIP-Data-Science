```python
import os
os.sys.path
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image=cv2.imread(r"C:\Users\hp\Pictures\Saved Pictures\testing.jpg")
image
```

```python
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,6))
plt.imshow(image_rgb)
plt.title("Original picture")
plt.axis("off")
```

```python
# convert image original into grayscale image

gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10,6))
plt.imshow(gray,cmap="gray")
plt.title("Gray scale Image")
plt.axis("off")
```

```python
# inverting image

inverted = 255-gray
plt.figure(figsize=(10,6))
plt.imshow(inverted,cmap="gray")
plt.title("Inverted image")
plt.axis("off")
```

```python
# blur image

blur = cv2.GaussianBlur(inverted,(21,21),0)
plt.figure(figsize=(10,6))
plt.imshow(blur,cmap="gray")
plt.title("Blur image")
plt.axis("off")
```

```python
# convert image into pencil sketch 

sketch = cv2.divide(gray,255-blur, scale=255)
plt.figure(figsize=(10,6))
plt.imshow(sketch,cmap="gray")
plt.title("Pencil sketch of image")
plt.axis("off")
```
